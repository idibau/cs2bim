## Convert WKT to TIN object

### Create ``polygon.Area`` object

```python
area = Area(wkt[2], origin=origin[:2])
```

### Load all raster points within perimeter (in this case the whole tile)

```python
dtm_points = RasterPoints(p_raster, origin=origin)
```

### Retrieve relevant raster points

```python
raster_points_buffer = dtm_points.within(area.get_geometry, buffer_dist= 3 * grid_size)
raster_points_within = dtm_points.within(area.get_geometry, buffer_dist= 0)
```

### Create initial mesh

```python
m = Mesh(raster_points_buffer)
```

### Clip mesh by Area

```python
m_clipped = m.clip_mesh_by_area(area, raster_points_within)
```

### Decimate number of triangles

```python
m_clipped_decimated = m_clipped.decimate(max_height_error=max_height_error, grid_size=grid_size)


print(f"# triangles original : {m_clipped.n_triangles}")
print(f"# triangles decimated : {m_clipped_decimated.n_triangles}")
```

### Area consistency check

To verify the conversion an area consistency check is provided. Thereby, the 2d area of the initial Area object is compared with the summed 2d areas of all triangle in the created surface object.

```python
m_clipped_decimated.check_area_consistency(area.get_area, treshold=0.1)
```

```python
print(f"Target 2D area: {np.round(area.get_area, 5)}")
print(f"2D area TIN: {np.round(m_clipped.get_area_2d, 5)}")
print(f"2D area TIN decimated: {np.round(m_clipped_decimated.get_area_2d, 5)}")
```
